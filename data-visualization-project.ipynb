```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df =pd.read_csv("/kaggle/input/disney-movies-19372016-total-gross/disney_movies_total_gross.csv")
df.head()
```

```python
inflation_adjusted_gross_desc = df.sort_values(by='inflation_adjusted_gross', ascending=False) 

# Display the top 10 movies 
inflation_adjusted_gross_desc.head(10)
sdf = inflation_adjusted_gross_desc.head(10)
fig, ax = plt.subplots(figsize=(10,10)) 
chart = sns.barplot(y=sdf.inflation_adjusted_gross, x=sdf.movie_title)
_=chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
plt.title('Top 10 inflation adjusted gross')
```

```python
genre_counts = df['genre'].value_counts()
plt.bar(genre_counts.index, genre_counts.values)
plt.title('Disney Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45) # Rotate x-axis labels by 45 degrees
plt.show()
```
